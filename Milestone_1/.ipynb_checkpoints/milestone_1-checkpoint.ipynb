# 3. Rozkład według zmiennej objaśnianej

```python
for column_name in [
    "History of Mental Illness", 
    "History of Substance Abuse", 
    "Family History of Depression",
    "Chronic Medical Conditions"
    ]:
    data[column_name] = (data[column_name] == "Yes")
```

```python
plt.figure(figsize=(20, 40))
for index in range(len(categorical_columns)):
    plt.subplot(len(categorical_columns), 2, index+1)
    sns.barplot(data=data.groupby(categorical_columns[index])["Chronic Medical Conditions"].mean().reset_index(), 
                x = categorical_columns[index],
                y="Chronic Medical Conditions"
               )
plt.show()
```

```python
sns.barplot(data=data.groupby("Number of Children")["Chronic Medical Conditions"].mean().reset_index(), x="Number of Children", y="Chronic Medical Conditions")
plt.show()
```

```python
plt.figure(figsize=(16, 14))
plt.subplot(2, 2, 1)
sns.histplot(data[data["Chronic Medical Conditions"]]["Income"], color="red")
plt.title("Depressed")
plt.subplot(2, 2, 2)
sns.histplot(data[~data["Chronic Medical Conditions"]]["Income"], color="green")
plt.title("Healthy")
plt.subplot(2, 2, 3)
sns.histplot(np.log(data[data["Chronic Medical Conditions"]]["Income"]), color="red")
plt.title("Log-Depressed")
plt.subplot(2, 2, 4)
sns.histplot(np.log(data[~data["Chronic Medical Conditions"]]["Income"]), color="green")
plt.title("Log-Healthy")
plt.show()
```

Na powyższych wykresach nie widać większej zależności

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 2, 1)
sns.histplot(data[data["Chronic Medical Conditions"]]["Age"], bins=np.linspace(18, 81, 64), color="red")
plt.title("Depressed")

plt.subplot(1, 2, 2)
sns.histplot(data[~data["Chronic Medical Conditions"]]["Age"], bins=np.linspace(18, 81, 64), color="green")

plt.title("Not-Depressed")
plt.show()
```

```python
plt.scatter(data=data.groupby("Age")["Chronic Medical Conditions"].mean().reset_index(), x="Age", y="Chronic Medical Conditions")
plt.ylim([0.25, 0.4])
plt.xlabel("Age")
plt.ylabel("Fraction of depressed")
plt.grid("on")
plt.show()
```

# 4. Badanie średniego poziomu depresji w zależności od kombinacji cech kategorycznych i wieku

```python
df = pd.read_csv("depression_data.csv")
df = pipeline(df)
categorical_columns = [colname for colname in df.columns if df[colname].dtype in [np.int8, np.int64, np.bool] if colname != "Chronic Medical Conditions"]
categorical_columns = np.array(categorical_columns)
all_combinations = [list(categorical_columns[[i, j]]) for i in range(len(categorical_columns)-1) for j in range(i+1, len(categorical_columns))]
all_combinations = [x for x in all_combinations if x[0] != x[1]]
```

```python
from matplotlib.colors import LinearSegmentedColormap
custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", ["green", "yellow", "red"])
plt.figure(figsize=(10, 140))
for i in range(len(all_combinations)):
    plt.subplot(int(np.ceil(len(all_combinations)/2)), 2, i+1)
    combination = all_combinations[i]
    x = df.groupby(combination)["Chronic Medical Conditions"].mean().reset_index()
    df_wide = x.pivot(index=combination[0], columns=combination[1], values="Chronic Medical Conditions")
    sns.heatmap(df_wide, cmap=custom_cmap)
    plt.xlabel(combination[1])
    plt.ylabel(combination[0])
```

```python
categorical_columns = [colname for colname in df.columns if df[colname].dtype in [np.int8, np.int64]]
categorical_columns = np.array(categorical_columns)
all_combinations = [list(categorical_columns[[i, j]]) for i in range(len(categorical_columns)-1) for j in range(i+1, len(categorical_columns))]
all_combinations = [x for x in all_combinations if x[0] != x[1]]
df["count"] = 1

custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", ["darkblue", "green", "yellow"])

plt.figure(figsize=(10, 140))
for i in range(len(all_combinations)):
    plt.subplot(int(np.ceil(len(all_combinations)/2)), 2, i+1)
    combination = all_combinations[i]
    x = df.groupby(combination).count().reset_index()
    df_wide = x.pivot(index=combination[0], columns=combination[1], values="count")
    sns.heatmap(df_wide, cmap=custom_cmap)
    #plt.title(f"{combination[0]}-{combination[1]}")
    plt.xlabel(combination[1])
    plt.ylabel(combination[0])
```
